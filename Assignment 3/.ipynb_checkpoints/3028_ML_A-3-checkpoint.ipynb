```python
X=data3[["GRE Score","TOEFL Score","University Rating","SOP","LOR ","CGPA","Research"]]
X
```

```python
Y=data3["Admitted"]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cl=DecisionTreeClassifier(criterion="gini")
```

```python
cl.fit(X_train,Y_train)
```

```python
Y_pred=cl.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay , accuracy_score , recall_score , precision_score
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test,Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
recall_score(Y_test,Y_pred)
```

```python
precision_score(Y_test,Y_pred)
```

```python
d=pd.DataFrame({"GRE Score":[305] , "TOEFL Score":[120] , "University Rating":[3] , "SOP":[2.7] , "LOR ":[3.6] , "CGPA":[8.06] , "Research":[0]})
```

```python
pr=cl.predict(d)
if(pr == 0) :
    print("Not Admitted")
else:
    print("Admitted")
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20, 14))
plot_tree(cl,feature_names=data3.columns.tolist(),filled=True,class_names=['Not Admitted', 'Admitted'])
plt.title("Decision Tree Classifier")
plt.show()
```
