```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
admission = pd.read_csv('Admission_Predict.csv')
```

```python
admission
```

Data PreProcessing

```python
ap = admission.drop(columns=['Serial No.'],axis=1)
```

```python
ap
```

```python
ap["Admitted"] = np.where(ap["Chance of Admit "]>=0.85, "Admitted","Not Admitted")
```

```python
ap
```

```python
ap = ap.drop(columns = ["Chance of Admit "])
```

```python
ap
```

```python
ap.isnull().sum()
```

```python
sb.countplot(x=ap['Admitted'])
```

```python
x = ap[['GRE Score','TOEFL Score','University Rating','SOP','LOR ','Research','CGPA' ]]
y = ap['Admitted']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,random_state=0,test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier(criterion="gini")
```

```python
clf1.fit(X_train, Y_train)
```

```python
Y_pred1 = clf1.predict(X_test)
```

```python
Y_pred1
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(Y_test,Y_pred1)
cm
```

```python
cm_display = metrics.ConfusionMatrixDisplay(cm, display_labels =[False, True] )
```

```python
cm_display.plot()
plt.show()
```

```python
Accuracy = metrics.accuracy_score(Y_test,Y_pred1)
Accuracy
```

#classification

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred1))
```
