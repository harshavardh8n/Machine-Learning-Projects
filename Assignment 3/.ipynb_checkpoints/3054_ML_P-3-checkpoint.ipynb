# Assignment No - 3 Classification Technique


```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
admission = pd.read_csv('Admission_Predict.csv')
```

```python
admission
```

# Data Preprocessing

```python
ap = admission.drop(columns = ["Serial No."], axis = 1)
ap
```

```python
ap['Admitted'] = np.where(ap["Chance of Admit "] >= 0.85, "Admitted", "Not Admitted") 
ap
```

```python
ap = ap.drop(columns = ["Chance of Admit "], axis = 1)
ap
```

```python
ap.isnull().sum()
```

```python
sb.countplot(x=ap["Admitted"])
```

```python
sb.barplot(data=ap,x="Admitted",y="CGPA",hue="University Rating")
```

```python
sb.scatterplot(data=ap,x="CGPA",y="GRE Score",hue="Admitted")
```

```python
plt.figure(figsize=(20, 14))
sb.boxplot(data=ap,x="GRE Score",y="TOEFL Score",hue="Admitted")
plt.plot()
```

```python
plt.figure(figsize=(20, 14))
sb.histplot(data=ap,x="GRE Score",hue="Research",palette="deep")
plt.plot()
```

```python
plt.figure(figsize=(20, 14))
count=ap["Admitted"].value_counts()
plt.pie(count,labels=count.index,autopct='%.1f%%')
plt.show()
```

```python
ap
```

```python
x = ap[['GRE Score','TOEFL Score','University Rating','SOP','LOR ','Research','CGPA']]
y = ap['Admitted']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1 = DecisionTreeClassifier(criterion= "gini")
clf1.fit(x_train, y_train)
```

```python
y_pred1 = clf1.predict(x_test)
y_pred1
```

```python
from sklearn import metrics 
confusion_matrix = metrics.confusion_matrix(y_test, y_pred1) 
confusion_matrix
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True]) 
```

```python
cm_display.plot()
plt.show() 
```

```python
Accuracy = metrics.accuracy_score(y_test, y_pred1)
Accuracy
```

# Classification

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred1))
```

```python
from sklearn import tree
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20, 14))
plot_tree(clf1,feature_names=ap.columns.tolist(),filled=True,class_names=['Not Admitted', 'Admitted'])
plt.title("Decision Tree Classifier")
plt.show()
```

```python
plt.figure(figsize=(20, 14))
plot_tree(clf1,feature_names=ap.columns.tolist(),filled=True,max_depth=2,class_names=['Not Admitted', 'Admitted'])
plt.title("Decision Tree Classifier")
plt.show()
```

```python
data ={"GRE Score":[322],"TOEFL Score":[110],"University Rating":[3],"SOP":[3.5],"LOR ":[2.5],"Research":[1],"CGPA":[8.67]}
frame=pd.DataFrame(data)
prediction=clf1.predict(frame)
prediction
```

```python
if(prediction==1):
    print("Student Admitted")
else:
    print("student not admitted")
```

```python
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier(criterion= 'entropy')
clf2.fit(x_train, y_train)
```

```python
y_pred2 = clf2.predict(x_test)
y_pred2 = pd.DataFrame(y_pred2, columns = ['Predicted'])
print("Admitted Value Predicted on the Basis of other Data :-")
y_pred2
```

```python
from sklearn import metrics 
confusion_matrix = metrics.confusion_matrix(y_test, y_pred2) 
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True]) 
```

```python
cm_display.plot()
plt.show() 
```

```python
Accuracy = metrics.accuracy_score(y_test, y_pred2)
Accuracy
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred2))
```
